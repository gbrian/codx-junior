```python
from playwright.async_api import async_playwright

pw = await async_playwright().start()
print("Lauching browser")
browser = await pw.chromium.launch_persistent_context(
    "/home/codxuser/user-data",
    channel="chrome",
    headless=False,
    args=[
        "--start-maximized"
    ],
    no_viewport=True
)
```

```python
await browser.pages[0].goto("https://vpnfg.w2m.com:10443/sslvpn/portal.html")
```

```python
await browser.close()
```
